```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------
# Inputs
# -------------------------
inputs = np.array([1.0, 2.0, 3.0])

# Weights
weights = np.array([0.5, 0.2, -0.1])


# =========================
# ReLU Function
# =========================
def relu(x):
    return np.maximum(0, x)


# =========================
# 1. Arbitrary Bias Model
# =========================
bias = 0.3   # change this value

net1 = np.dot(inputs, weights) + bias
output1 = relu(net1)

print("Arbitrary Bias Model (ReLU)")
print("Net value:", net1)
print("Output:", output1)


# =========================
# 2. Fixed Bias Model
# =========================
bias_fixed = 1   # fixed value

net2 = np.dot(inputs, weights) + bias_fixed
output2 = relu(net2)

print("\nFixed Bias Model (ReLU)")
print("Net value:", net2)
print("Output:", output2)


# =========================
# ReLU Activation Graph
# =========================
x_values = np.linspace(-5, 5, 100)
y_values = relu(x_values)

plt.plot(x_values, y_values)
plt.title("ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```
